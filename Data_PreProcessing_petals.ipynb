```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris_data = load_iris()

data = pd.DataFrame(iris_data.data, columns = iris_data.feature_names)
data
```

```python
data.columns
```

```python
data.shape
```

```python
data['sepal length (cm)'].describe()
```

```python
print('Skewness : ', data['sepal length (cm)'].skew())
print('Kurtosis : ', data['sepal length (cm)'].kurt())
```

```python
sns.distplot(data['sepal length (cm)'])
```

```python
var='sepal width (cm)'
data.plot.scatter(var,'sepal length (cm)')
```

```python
data[['sepal length (cm)', 'sepal width (cm)']].corr()
```

```python
f, ax = plt.subplots(figsize = (12,9))
sns.heatmap(data.corr(), square=True, annot=True)
```

```python
#scatterplot
cols= ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)',
       'petal width (cm)']
sns.pairplot(data[cols]);
```
